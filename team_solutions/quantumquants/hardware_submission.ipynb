# Submitting Your Program to *Aquila*

```julia
from bloqade.atom_arrangement import ListOfLocations
```

```julia
positions = [
    (0, 0),
    (1, 0),
    (2, 0),
    (4, 0),
    (2, 1),
    (3, 1),
    (0, 2),
    (3, 2),
    (4, 2),
    (3, 3),
    (4, 3),
    (0, 4),
    (2, 4),
    (3, 4),
    (4, 4)
]

list_pos = [(pos[0]*4, pos[1]*4) for pos in positions]

omega_durations = [0.2, 3.6, 0.2]
omega_values = [0, 6.544289568279506, 6.544289568279506, 0.0]

delta_durations = [0.2, 3.24, 0.36, 0.2]
delta_values = [-27.971415389660272, -27.971415389660272, 23.594454699681336, 16.185740815456843, 16.185740815456843]

mis_udg_program = (
    ListOfLocations(list_pos)
    .rydberg.rabi.amplitude.uniform.piecewise_linear(omega_durations, omega_values)
    .detuning.uniform.piecewise_linear(
        delta_durations, delta_values
    )
)

def generate_lattice(positions, lattice_size):
    list_pos = [(pos[0]*lattice_size, pos[1]*lattice_size) for pos in positions]

mis_udg_job = mis_udg_program #.assign(final_detuning=5)
```

## Submitting the Program

```julia
from braket.devices import Devices
from braket.jobs import hybrid_job
```

```julia
%env AWS_ACCESS_KEY_ID=
%env AWS_SECRET_ACCESS_KEY=
%env AWS_DEFAULT_REGION=
```

```julia
@hybrid_job(device=Devices.QuEra.Aquila, dependencies="./reqs.txt")
def your_job():
    return mis_udg_job.braket.aquila().run(100).report()
```

```julia
cloud_job_100 = your_job()
```

```julia
cloud_job_100.queue_position()
```
